```python
import pandas as pd

data = {
    "text": [
        "I love Lalettan movies",
        "Especially romantic Genre",
        "I am Excited today",
        "I hate Action movies",
        "The movie is good"
    ]
}

df = pd.DataFrame(data)
df
```

```python
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

df['score'] = df['text'].apply(lambda x: sia.polarity_scores(x)['compound'])
df['sentiment'] = df['score'].apply(lambda x:
                                    'Positive' if x > 0
                                    else ('Negative' if x < 0
                                    else 'Neutral'))
df
```

```python
from textblob import TextBlob

df['textblob_score'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['textblob_sentiment'] = df['textblob_score'].apply(
    lambda x: 'Positive' if x>0 else ('Negative' if x<0 else 'Neutral')
)
df
```

```python
texts = [
    "I LOVE Biriyani no neautiful!!! ",
    "Rice and sambar so boring  ",
    "After biriyan had a ice cream is exciting ",
    "Not sure if I like payasam. ",
    "Great food yummy "
]
```

```python
import pandas as pd
df = pd.DataFrame(texts, columns=["text"])
#df df['clean_text'] = df['text'].str.lower()
df['clean_text'] = df['text'].str.lower().str.strip()
df[['text','clean_text']]
```

converting text to lowercase is one of the most common text preprocessing steps in Natural Language Processing (NLP) and text analysis.

```python
import string
df['no_punct'] = df['clean_text'].str.translate(str.maketrans('', '', string.punctuation))
df[['clean_text','no_punct']]
```

Removing punctuation can cause loss of structure, tone, and clarity  especially in tasks like sentiment analysis, chatbot design, or translation.

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop = set(stopwords.words('english'))
df['no_stopwords'] = df['no_punct'].apply(lambda x: " ".join([word for word in x.split() if word not in stop]))
df[['no_punct','no_stopwords']]
```

Removing stopwords is one of the most common text preprocessing steps, but its done for specific reasons depending on the task.

```python
import re

def remove_emojis(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"
                           u"\U0001F300-\U0001F5FF"
                           u"\U0001F680-\U0001F6FF"
                           u"\U0001F1E0-\U0001F1FF"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

df['no_emoji'] = df['no_stopwords'].apply(remove_emojis)
df[['no_stopwords','no_emoji']]
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from textblob import TextBlob

# Initialize analyzers
vader = SentimentIntensityAnalyzer()

sentences = [
    "Oh great, another assignment ",
    "I just love waiting in traffic for hours "
]

for s in sentences:
    print(f"\nSentence: {s}")
    print("VADER:", vader.polarity_scores(s)['compound'])
    print("TextBlob:", TextBlob(s).sentiment.polarity)
```
