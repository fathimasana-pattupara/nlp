```python
# Example of how this code might be used in a loop
from textblob import TextBlob

reviews = ["This is a great product!", "It was okay.", "I did not like it at all."]
sentiments = []
polarities = []
categories = []

for review in reviews:
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity

    if polarity > 0:
        category = "Positive"
    elif polarity < 0:
        category = "Negative"
    else:
        category = "Neutral"

    sentiments.append(review)
    polarities.append(polarity)
    categories.append(category)

# Now you can work with the populated lists
print("Reviews:", sentiments)
print("Polarities:", polarities)
print("Categories:", categories)
```
